```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import dash_core_components
import plotly.figure_factory as ff
import plotly.graph_objs as go
from plotly.offline import *
```

```python
df = pd.read_excel("Data_Cortex_Nuclear.xls")
df.head()
```

```python
df = df[["pPKCG_N", "pP70S6_N", "pS6_N", "pGSK3B_N", "ARC_N","class"]]
df.head()
```

```python
df = df[(df["class"] == 't-CS-s') | (df["class"] == 'c-CS-s')]
```

```python
fig = ff.create_scatterplotmatrix(df, diag='histogram', index='class', height=1000, width=1000)

iplot(fig)
```
